```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
grid_size = 200
x = np.linspace(-10, 10, grid_size)
y = np.linspace(-10, 10, grid_size)
X, Y = np.meshgrid(x, y)

# Define the polar coordinates
R = np.sqrt(X**2 + Y**2)
Theta = np.arctan2(Y, X)

# Define the phase of the wavefunction for n=1 vortex
n = 1  # Winding number
phase = n * Theta

# Define the amplitude of the wavefunction (zero at the vortex core)
amplitude = np.exp(-R**2 / 10)  # Decay away from the vortex core

# Combine amplitude and phase to create the wavefunction
wavefunction = amplitude * np.exp(1j * phase)

# Plot the phase of the wavefunction
plt.figure(figsize=(8, 8))
plt.pcolormesh(X, Y, np.angle(wavefunction), cmap='hsv', shading='auto')
plt.colorbar(label="Phase (radians)")
plt.title("Phase of n=1 Vortex")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()

# Plot the amplitude of the wavefunction
plt.figure(figsize=(8, 8))
plt.pcolormesh(X, Y, np.abs(wavefunction), cmap='viridis', shading='auto')
plt.colorbar(label="Amplitude")
plt.title("Amplitude of n=1 Vortex")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D

# Create a 3D grid for the vortex model
grid_size = 200
x = np.linspace(-10, 10, grid_size)
z = np.linspace(-10, 10, grid_size)
X, Z = np.meshgrid(x, z)
Y = np.sqrt(X**2 + Z**2)  # Create a radial symmetry in 3D

# Define the amplitude and phase for the vortex
R = np.sqrt(X**2 + Z**2)
phase_3d = np.arctan2(Z, X)
amplitude_3d = np.exp(-R**2 / 10)  # Amplitude decays with radius

# Create the 3D vortex by modulating amplitude with phase
vortex = amplitude_3d * np.sin(phase_3d)

# Plot the 3D vortex
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Z, vortex, cmap='hsv', edgecolor='none', alpha=0.8)
ax.set_title("3D Vortex Model (n=1)", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Z")
ax.set_zlabel("Amplitude")
ax.view_init(30, 60)  # Adjust the viewing angle for clarity
plt.show()
```
